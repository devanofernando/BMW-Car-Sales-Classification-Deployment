## Import Libraries

```python
import pandas as pd
import joblib
```

## Load Model Terbaik

```python
model = joblib.load('best_bmw_sales_model.pkl')
```

## New Data

```python
# Masukkan data baru
new_data = pd.DataFrame([{
    'Model': 'X1',
    'Year': 2023,
    'Region': 'Asia',
    'Color': 'Black',
    'Fuel_Type': 'Petrol',
    'Transmission': 'Automatic',
    'Engine_Size_L': 3.0,
    'Mileage_KM': 15000,
    'Price_USD': 85000,
    'Sales_Volume': 5000
}])

print("Data baru:")
display(new_data)
```

## Predict Data

```python
# Lakukan prediksi
prediction = model.predict(new_data)
prediction_proba = model.predict_proba(new_data)

# Tampilkan hasil
print(f"\nPrediksi Sales Classification: {'High' if prediction[0] == 1 else 'Low'}")
print(f"Probabilitas: Low = {prediction_proba[0][0]:.2%}, High = {prediction_proba[0][1]:.2%}")
```

## Kesimpulan

Model inference sudah dilakukan dengan melakuakn penambahan data yang baru. Hasil dari prediksi mengatakan bahwa data yang baru memiliki sales classification kategori 'Low' dengan probability low = 100% (tidak ada probability high)
